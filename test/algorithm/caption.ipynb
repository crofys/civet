```python
class Encoder(Model):
  def __init__(self):
    super(Encoder, self).__init__()
    resnet = ResNet50()
    modules = resnet.layers[:-2]
#     print(modules)
    self.resnet = Sequential([modules[:5] ])

  def call(self, images):
    out = self.resnet(images)
    return out

model = Encoder()
```
